# **Problem 1: Mobile Recharge Validation System**

```python
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        print("Recharge amount must be at least ₹50")
        return False
    elif amount not in valid_plans:
        print("Invalid plan. Valid plans are:", valid_plans)
        return False
    else:
        print("Recharge successful for ₹", amount)
        return True


while True:
    amount = int(input("Enter recharge amount: ₹"))

    if validate_recharge(amount):
        break
    else:
        print("Please try again.\n")
```

# **Problem 2: Inventory Reorder Alert System**

```python
def check_inventory(stock):
    for product, quantity in stock.items():
        if quantity < 15:
            print(product, ": Reorder Alert")
        else:
            print(product, ": Stock OK")


# Sample inventory data
inventory = {
    "Rice": 10,
    "Sugar": 25,
    "Oil": 8,
    "Wheat": 20
}

check_inventory(inventory)
```

# **Problem 3: Student Result Processing System**

```python
def student_result(marks):
    total = 0

    for mark in marks:
        total += mark

    average = total / len(marks)

    if average >= 50:
        return "Pass"
    else:
        return "Fail"

n = int(input("Enter number of subjects: "))

marks_list = []
for i in range(n):
    mark = int(input(f"Enter marks for subject {i+1}: "))
    marks_list.append(mark)

#Result
result = student_result(marks_list)
print("Result:", result)
```

# **Problem 4: Cab Fare Estimator with Retry Option**

```python
def calculate_fare(distance, is_peak):
    base_fare = 50
    per_km_rate = 12

    total = base_fare + (distance * per_km_rate)

    if is_peak:
        total += total * 0.25  # 25% extra

    return total


#Retry
while True:
    distance = float(input("Enter distance in km: "))
    peak = input("Is it peak hour? (yes/no): ").lower()

    is_peak = True if peak == "yes" else False

    fare = calculate_fare(distance, is_peak)
    print(f"Total Cab Fare: ₹{fare:.2f}")

    retry = input("Do you want to calculate again? (yes/no): ").lower()
    if retry != "yes":
        break
```

# **Problem 5: Employee Attendance Eligibility Checker**

```python
def check_attendance(attendance):
    present_count = 0
    for day in attendance:
        if day == "P":
            present_count += 1
    percentage = (present_count / len(attendance)) * 100
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

n = int(input("Enter total number of working days: "))

attendance_list = []
for i in range(n):
    day = input(f"Enter attendance for day {i+1} (P/A): ").upper()
    attendance_list.append(day)
#Result
result = check_attendance(attendance_list)
print("Attendance Status:", result)
```

# **Problem 6: Password Strength Checker**


```python
def check_password(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"

    #Check length
    if len(password) < 8:
        return "Weak Password"

    #iterating through password characters
    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in special_chars:
            has_special = True

    #check
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"

pwd = input("Enter password: ")
result = check_password(pwd)
print(result)
```
