Problem 1: Smart Parking Lot Management System

```python
def smart_parking(capacity, logs):
    parked = 0

    # Count vehicles entering and exiting
    for log in logs:
        if log == "IN":
            parked += 1
        elif log == "OUT":
            parked -= 1

    print("Currently Parked Vehicles:", parked)

    # Check parking status
    if parked > capacity:
        print("Alert: Parking Exceeded Capacity")
    else:
        print("Parking Status: Available")


# Sample Input
capacity = 50
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

smart_parking(capacity, logs)
```

Problem 2: Online Food Delivery Time Estimator

```python
def delivery_time(distance, traffic, weather):

    # Base time (5 minutes per km)
    time = distance * 5

    # Traffic delay
    if traffic == "High":
        time += 10
    elif traffic == "Medium":
        time += 5

    # Weather delay
    if weather == "Rainy":
        time += 5
    elif weather == "Storm":
        time += 10

    print("Estimated Delivery Time:", time, "minutes")


# Sample Input
delivery_time(8, "High", "Rainy")
```

Problem 3: Movie Theatre Seat Occupancy Analyzer

```python
def theatre_occupancy(total_seats, booked_list):

    booked = len(booked_list)
    occupancy = (booked / total_seats) * 100

    print("Occupancy:", int(occupancy), "%")

    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")


# Sample Input
total_seats = 200
booked_seats = [1]*150

theatre_occupancy(total_seats, booked_seats)
```

Problem 4: Cloud Server Load Classification System

```python
def server_load(cpu_readings):

    avg = sum(cpu_readings) / len(cpu_readings)

    print("Average CPU Load:", int(avg), "%")

    if avg < 50:
        print("Server Status: Normal")
    elif avg <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


# Sample Input
cpu = [45, 60, 70, 85, 90]

server_load(cpu)
```

Problem 5: Smart Classroom Resource Usage Monitor

```python
def resource_monitor(resources):

    overused = []

    for resource, hours in resources.items():
        if hours > 8:
            overused.append(resource)

    if overused:
        print("Overused Resources:", ",".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# Sample Input
usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

resource_monitor(usage)
```

Problem 6: Online Event Registration Capacity Controller

```python
def event_registration(capacity, registrations):

    if registrations > capacity:
        confirmed = capacity
        waitlist = registrations - capacity
        status = "Closed"
    else:
        confirmed = registrations
        waitlist = 0
        status = "Open"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlist)
    print("Registration Status:", status)


# Sample Input
event_registration(100, 105)
```
