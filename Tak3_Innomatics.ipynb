
Problem Statement 1: Employee Performance Bonus Eligibility


```python
# Take number of employees
n = int(input("Enter number of employees: "))

employees = {}

# Input employee data
for _ in range(n):
    name = input("Enter employee name: ")
    score = int(input("Enter performance score: "))
    employees[name] = score

# Find highest score
highest_score = max(employees.values())

# Find top performers
top_performers = [name for name, score in employees.items() if score == highest_score]

print(f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (Score: {highest_score})")
```

Problem Statement 2: Search Query Keyword Analysis

```python
import string

# Take search query from user
query = input("Enter search query: ")

# Convert to lowercase
query = query.lower()

# Remove punctuation
query = query.translate(str.maketrans('', '', string.punctuation))

# Split into words
words = query.split()

# Count word frequency
word_count = {}
for word in words:
    word_count[word] = word_count.get(word, 0) + 1

# Filter words searched more than once
result = {word: count for word, count in word_count.items() if count > 1}

print(result)
```

Problem Statement 3: Sensor Data Validation

```python
# Take sensor readings from user
sensor_readings = list(map(int, input("Enter sensor readings separated by space: ").split()))

# Store even readings with index
valid_readings = [(index, value) for index, value in enumerate(sensor_readings) if value % 2 == 0]

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

Problem Statement 4: Email Domain Usage Analysis


```python
# Take number of emails
n = int(input("Enter number of email IDs: "))

emails = []

# Input emails
for _ in range(n):
    emails.append(input("Enter email ID: "))

# Extract domains
domain_count = {}

for email in emails:
    domain = email.split('@')[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1

# Calculate and display percentage usage
total_emails = len(emails)

for domain, count in domain_count.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {int(percentage)}%")
```

Problem Statement 5: Sales Spike Detection (User Input)

```python
# Take daily sales data
sales = list(map(int, input("Enter daily sales values separated by space: ").split()))

# Calculate average sales
average_sales = sum(sales) / len(sales)

# Threshold = 30% above average
threshold = average_sales * 1.3

for day, value in enumerate(sales, start=1):
    if value > threshold:
        print(f"Day {day}: {value}")
```

Problem Statement 6: Duplicate User ID Detection

```python
# Given list of user IDs
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Dictionary to store frequency of each user ID
user_count = {}

# Count occurrences of each user ID
for user in user_ids:
    user_count[user] = user_count.get(user, 0) + 1

# Display duplicate user IDs and their counts
for user, count in user_count.items():
    if count > 1:
        print(f"{user} → {count} times")
```
