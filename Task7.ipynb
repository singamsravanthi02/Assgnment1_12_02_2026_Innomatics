1. Smart Light Controller

```python
# Class to control smart light
class SmartLight:
    def __init__(self, light_name):
        self.light_name = light_name

    def turn_on(self):
        print(self.light_name + " is ON")

    def turn_off(self):
        print(self.light_name + " is OFF")


# Input
name = input("Light Name: ")
action = input("Action (ON/OFF): ")

light = SmartLight(name)

if action.upper() == "ON":
    light.turn_on()
else:
    light.turn_off()
```

2. Employee ID Card System

```python
# Class to store employee details
class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_card(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)


# Input
name = input("Employee Name: ")
emp_id = input("Employee ID: ")
dept = input("Department: ")

emp = Employee(name, emp_id, dept)
emp.display_card()
```

3. Mobile Contact Record

```python
# Class to store contact details
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


# Input
name = input("Contact Name: ")
phone = input("Phone Number: ")

c = Contact(name, phone)
c.display_contact()
```

4. Product Price Tag Generator

```python
# Class to store product details
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def price_tag(self):
        print("Product:", self.name)
        print("Price: ₹" + str(self.price))


# Input
name = input("Product Name: ")
price = input("Price: ")

p = Product(name, price)
p.price_tag()
```

5. Movie Rating Display System

```python
# Class to store movie details
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display_movie(self):
        print("Movie:", self.name)
        print("Rating:", self.rating, "/ 5")


# Input
name = input("Movie Name: ")
rating = input("Rating: ")

m = Movie(name, rating)
m.display_movie()
```
