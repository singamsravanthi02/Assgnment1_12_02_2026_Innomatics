1. User Login Check
Problem Statement:
Given a username and password, check whether login is successful.
username = "admin"
password = "1234"

Requirements:
Print "Login Successful" if both username and password match
Otherwise print "Invalid Credentials"
Real-World Application: Authentication systems


```python
def CheckLogin(username, password):
    if username == "admin" and password == "1234":
        return "Login Successful"
    else:
        return "Invalid Credentials"

user_name = input("Enter username: ")
user_pass = input("Enter password: ")

print(CheckLogin(user_name, user_pass))
```

2. Pass / Fail Analyzer
Problem Statement:
Analyze student results.
marks = [45, 78, 90, 33, 60]

Requirements:
A student passes if marks ≥ 50
Count the total number of pass students
Count the total number of fail students
Print the final result clearly
Real-World Application: Academic evaluation systems


```python
def analyzeResults(marks):
    pass_count = 0
    fail_count = 0

    for mark in marks:
        if mark >= 50:
            pass_count += 1
        else:
            fail_count += 1

    print(f"Total Pass students: {pass_count}")
    print(f"Total Fail students: {fail_count}")
marks = [45, 78, 90, 33, 60]
analyzeResults(marks)
```

3. Simple Data Cleaner
Problem Statement:
Clean and standardize user names.
names = [" Alice ", "bob", " CHARLIE "]

Requirements:
Remove extra spaces
Convert all names to lowercase
Print the cleaned list
Real-World Application: Data preprocessing before analysis


```python
def cleanNames(names):
    cleaned = [name.strip().lower() for name in names]
    print("Cleaned names:", cleaned)
names = [" Alice ", "bob", " CHARLIE "]
cleanNames(names)
```

4. Message Length Analyzer
Problem Statement:
Analyze message sizes.
messages = ["Hi", "Welcome to the platform", "OK"]

Requirements:
Print the length of each message
Flag messages longer than 10 characters
Real-World Application: Text filtering and validation systems


```python
def analyzeMessages(messages):
    print("Message Lengths:")
    for msg in messages:
        length = len(msg)
        status = "LONG" if length > 10 else "OK"
        print(f"'{msg}' → {length} chars ({status})")

# Given messages
messages = ["Hi", "Welcome to the platform", "OK"]
analyzeMessages(messages)
```

5. Error Message Detector
Problem Statement:
Detect error messages from system logs.
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

Requirements:
Count the number of "ERROR" entries
Print the total error count
Real-World Application: Monitoring and log analysis systems

```python
def analyzeLogs(logs):
    """
    Analyzes system logs and counts ERROR entries.
    Real-world use: Monitoring/alerting systems.
    """
    error_count = 0

    # Count errors by checking each log entry
    for log in logs:
        if log == "ERROR":
            error_count += 1

    # Print result clearly (no f-strings)
    print("Total ERROR entries: {}".format(error_count))
    print("Log analysis complete!")

# Given logs
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
analyzeLogs(logs)
```
